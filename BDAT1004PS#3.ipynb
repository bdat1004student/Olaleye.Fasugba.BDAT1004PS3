```python
 # Reading dataset and printing head
link = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
#Assign it to a variable called users
users = pd.read_csv(link, sep = "|")
users.head()
```

```python
 # Mean age per occupation
users.groupby('occupation').age.mean()
```

```python
 # Function or filter for getting only males
def gender(var):
    if var == 'M':
        return True
    else:
        return False
```

```python
 # Applying filter for male gender
users['maleOnly'] = users['gender'].apply(gender)
#occupation by males
maleNo = users.groupby('occupation')['maleOnly'].sum()
#occupation by both genders
totalNo = users.groupby('occupation').gender.count()
 # Finding  male ratio and sorting in descending order
ratio = maleNo/totalNo
ratioSort = ratio.sort_values(ascending = False)
ratioSort
```

```python
 # For each occupation, calculating the minimum age

users.groupby('occupation').age.min()
```

```python
 # For each occupation, calculating the maximum age
users.groupby('occupation').age.max()
```

```python
 # For each combination of occupation and sex, calculate the mean age 
users.groupby([ 'occupation' , 'gender' ])[ 'age' ].mean ()
```

```python
# For each occupation calculating the percentage of women and men 
# Counting each gender in each occupation
genderCount= users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
# Calculating total people in each occupation
totalCount = users.groupby(['occupation']).count()
# Calculating the percentage by each gender
percentageBygender = genderCount.div(totalCount, level = "occupation") * 100
percentageBygender.loc[:, 'gender']
```

```python
 # Libraries
import pandas as pd
import datetime
```

```python
 # Reading dataset, applying first 3 columns to proper date time frame and printing values
data=pd.read_csv('wind.data.txt',sep="\s+",parse_dates=[[0,1,2]])

data.head()
```

```python
# Data from 2061
# Function fto fix years
def fixYears(var):
    if var.year > 1979:
        year=var.year-100
    else:
        year = var.year
    return datetime.date(year,var.month,var.day)

```

```python
 # Applying function and printng values
data['Yr_Mo_Dy']=data['Yr_Mo_Dy'].apply(fixYears)
data.head()
```

```python
 # Setting right index for date
data['Yr_Mo_Dy']=pd.to_datetime(data['Yr_Mo_Dy'])
data=data.set_index('Yr_Mo_Dy')
data.head()
```

```python
 # Missing values count
data.isnull().sum()
```

```python
 # Non missing values
data.shape[0]-data.isnull().sum() 
```

```python
 # Mean wind speed
data.mean().mean() 
```

```python
 # Creating datafram loc_stats and computing min, max, mean and std of windspeeds over all days
loc_stats=pd.DataFrame()
loc_stats['min']=data.min()
loc_stats['max']=data.max()
loc_stats['mean']=data.mean()
loc_stats['std']=data.std()
loc_stats
```

```python
 # Creating datafram day_stats and computing min, max, mean and std of windspeeds
day_stats=pd.DataFrame()
day_stats['min']=data.min(axis=1)
day_stats['max']=data.max(axis=1)
day_stats['mean']=data.mean(axis=1)
day_stats['std']=data.std(axis=1)
day_stats
```

```python
 # Mean windspeed in January
data['date']=data.index
data['month']=data['date'].apply(lambda date:date.month)
data['year']=data['date'].apply(lambda date:date.year)
data['day']=data['date'].apply(lambda date:date.day)
janWinds=data.query('month==1')
janWinds
janWinds.loc[:,'RPT':'MAL'].mean()
```

```python
 # Yearly frequency
data.query('month == 1 and day == 1') 
```

```python
 # Monthly frequency
data.query('day == 1')
```

```python
 # Weekly frequency
data.resample('W').mean()
```

```python
#Step 15. 
data1961 = data[data.index < pd.to_datetime('1962-01-01')]
data1961.resample('W').mean()
data1961.resample('W').min()
data1961.resample('W').max()
data1961.resample('W').std()
```

# Question 7

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
 # Loading data
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data.head()
```

```python
data_bw_3_years = data[data['Year'].apply(lambda x: x in [1900, 1950, 2000])]
marriages = data_bw_3_years['Marriages_per_1000'].values
divorces = data_bw_3_years['Divorces_per_1000'].values
```

```python
X = ['1900', '1950', '2000']
X_axis = np.arange(len(X))
plt.bar(X_axis - 0.2, marriages, 0.4, label = 'Marriages')
plt.bar(X_axis + 0.2, divorces, 0.4, label = 'Divorces')
plt.xticks(X_axis, X)
plt.xlabel("Years")
plt.ylabel("Values")
plt.legend()
 
# Show graph
plt.show()
```
